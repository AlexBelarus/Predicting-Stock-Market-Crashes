## Predicting Crashes in Financial Markets -- Comparison of Results
- Linear Regression / Logistic Regression / SVM (linear) / SVM (nonlinear) / Decision Trees
- 16 features vs 504 features
- Predictions for crash withinin 1 month / 3 months / 6 months

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
##### Print results 1 month prediction
d = {}
d['Linear Reg (16)'] = [0.12, 0.35, 0.94, 35, 0.13, 0.26, 0.94, 29]
d['Logistic Reg (16)'] = [0.2, 0.19, 0.97, 33, 0.17, 0.21, 0.96, 33]
d['SVC linear (16)'] = [0.06, 0.49, 0.86, 32, 0.06, 0.48, 0.8, 28]
d['SVC RBF (16)'] = [0.08, 0.33, 0.92, 28, 0.07, 0.4, 0.87, 28]
d['Decision Tree (16)'] = [1, 1, 1, 100, 0.07, 0.08, 0.95, 10]
results_1m = pd.DataFrame.from_dict(d, orient='index')
results_1m.columns = ['precision_tr', 'recall_tr', 'accuracy_tr', 'score_tr', \
                   'precision_t', 'recall_t', 'accuracy_t', 'score_t']

d = {}
d['Linear Reg (504)'] = [0.14, 0.4, 0.95, 40, 0.14, 0.29, 0.94, 33]
d['Logistic Reg (504)'] = [0.11, 0.33, 0.94, 32, 0.08, 0.31, 0.9, 26]
d['Decision Tree (504)'] = [1, 1, 1, 100, 0.09, 0.09, 0.96, 14]
results_1m_504 = pd.DataFrame.from_dict(d, orient='index')
results_1m_504.columns = ['precision_tr', 'recall_tr', 'accuracy_tr', 'score_tr', \
                   'precision_t', 'recall_t', 'accuracy_t', 'score_t']
print('Results 1m prediction with 16 features:')
display(results_1m)
print('Results 1m prediction with 504 features:')
display(results_1m_504)
```

```python
##### Plot results 1 month prediction
x = list(results_1m['recall_t'])
x_504 = list(results_1m_504['recall_t'])
y = list(results_1m['precision_t'])
y_504 = list(results_1m_504['precision_t'])
color = ['orange', 'red', 'blue', 'darkblue', 'green']
color_504 = ['orange', 'red', 'green']
rcParams['figure.figsize'] = 12, 8
plt.scatter(x=x, y=y, c=color, linewidths=5, alpha=0.8, marker='o')
plt.scatter(x=x_504, y=y_504, c=color_504, linewidths=5, alpha=0.8, marker='v')
labels = list(results_1m.index) + list(results_1m_504.index)
for i, txt in enumerate(results_1m.index):
    plt.annotate(txt, (x[i], y[i]), xytext=(10,-5), textcoords='offset points')
for i, txt in enumerate(results_1m_504.index):
    plt.annotate(txt, (x_504[i], y_504[i]), xytext=(10,-5), textcoords='offset points')
plt.xlim(xmin=0, xmax=0.7)
plt.ylim(ymin=0, ymax=0.3)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Crash within 1 month - Recall vs Precision')
plt.grid()
plt.show()
```
