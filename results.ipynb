## Predicting Crashes in Financial Markets -- Comparison of Results
- Linear Regression / Logistic Regression / SVM (linear) / SVM (nonlinear) / Decision Trees
- 16 features (for all models) vs 504 features (for regression models)
- Predictions for crash withinin 1 month / 3 months / 6 months

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
# 1 month prediction results:
d = {}
d['Linear Reg (16)'] = [0.12, 0.37, 0.95, 0.26, 0.14, 0.41, 0.95, 0.28]
d['Linear Reg (504)'] = [0.14, 0.42, 0.95, 0.3, 0.15, 0.4, 0.94, 0.28]
d['Logistic Reg (16)'] = [0.13, 0.35, 0.95, 0.26, 0.13, 0.41, 0.93, 0.25]
d['Logistic Reg (504)'] = [0.11, 0.53, 0.92, 0.3, 0.08, 0.42, 0.89, 0.21]
d['SVC lin'] = [0.15, 0.29, 0.96, 0.25, 0.15, 0.34, 0.95, 0.25]
d['SVC RBF'] = [0.16, 0.27, 0.96, 0.25, 0.16, 0.33, 0.95, 0.25]
d['Decision Tree'] = [1, 1, 1, 1, 0.09, 0.08, 0.96, 0.08]
results_1m = pd.DataFrame.from_dict(d, orient='index')
results_1m.columns = ['precision_tr', 'recall_tr', 'accuracy_tr', 'score_tr', \
                   'precision_t', 'recall_t', 'accuracy_t', 'score_t']
# 3 months prediction results:
d = {}
d['Linear Reg (16)'] = [0.15, 0.34, 0.89, 0.27, 0.17, 0.39, 0.89, 0.3]
d['Linear Reg (504)'] = [0.16, 0.37, 0.89, 0.3, 0.19, 0.39, 0.89, 0.31]
d['Logistic Reg (16)'] = [0.15, 0.35, 0.88, 0.27, 0.14, 0.42, 0.86, 0.27]
d['Logistic Reg (504)'] = [0.15, 0.41, 0.87, 0.3, 0.14, 0.44, 0.84, 0.28]
d['SVC lin'] = [0.17, 0.3, 0.9, 0.26, 0.16, 0.36, 0.89, 0.26]
d['SVC RBF'] = [0.17, 0.28, 0.91, 0.25, 0.16, 0.36, 0.89, 0.26]
d['Decision Tree'] = [1, 1, 1, 1, 0.09, 0.09, 0.91, 0.09]
results_3m = pd.DataFrame.from_dict(d, orient='index')
results_3m.columns = ['precision_tr', 'recall_tr', 'accuracy_tr', 'score_tr', \
                   'precision_t', 'recall_t', 'accuracy_t', 'score_t']
# 6 months prediction results:
d = {}
d['Linear Reg (16)'] = [0.17, 0.31, 0.82, 0.27, 0.22, 0.4, 0.83, 0.33]
d['Linear Reg (504)'] = [0.18, 0.33, 0.82, 0.29, 0.24, 0.42, 0.83, 0.35]
d['Logistic Reg (16)'] = [0.15, 0.41, 0.76, 0.3, 0.16, 0.5, 0.71, 0.31]
d['Logistic Reg (504)'] = [0.17, 0.38, 0.80, 0.3, 0.16, 0.47, 0.76, 0.32]
d['SVC lin'] = [0.2, 0.27, 0.85, 0.25, 0.2, 0.39, 0.83, 0.29]
d['SVC RBF'] = [0.21, 0.25, 0.86, 0.24, 0.21, 0.39, 0.84, 0.3]
d['Decision Tree'] = [1, 1, 1, 1, 0.13, 0.15, 0.84, 0.14]
results_6m = pd.DataFrame.from_dict(d, orient='index')
results_6m.columns = ['precision_tr', 'recall_tr', 'accuracy_tr', 'score_tr', \
                   'precision_t', 'recall_t', 'accuracy_t', 'score_t']

# Plot results recall vs precision:
x = list(results_1m['recall_t'])
y = list(results_1m['precision_t'])
color = ['orange','orange', 'red', 'red', 'blue', 'darkblue', 'green']
marker=['o','v','o','v','o','o','o']
rcParams['figure.figsize'] = 8, 18
plt.subplot(3,1,1)
for x_, y_, c_, m_ in zip(x, y, color, marker):
    plt.scatter(x_, y_, marker=m_, c=c_, linewidths=6, alpha=0.7)
plt.legend(results_1m.index)
plt.xlim(xmin=0, xmax=0.7)
plt.ylim(ymin=0, ymax=0.3)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Crash within 1 month - Recall vs Precision')
plt.grid()
plt.subplot(3,1,2)
x = list(results_3m['recall_t'])
y = list(results_3m['precision_t'])
for x_, y_, c_, m_ in zip(x, y, color, marker):
    plt.scatter(x_, y_, marker=m_, c=c_, linewidths=6, alpha=0.7)
plt.legend(results_3m.index)
plt.xlim(xmin=0, xmax=0.7)
plt.ylim(ymin=0, ymax=0.3)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Crash within 3 month - Recall vs Precision')
plt.grid()
plt.subplot(3,1,3)
x = list(results_6m['recall_t'])
y = list(results_6m['precision_t'])
for x_, y_, c_, m_ in zip(x, y, color, marker):
    plt.scatter(x_, y_, marker=m_, c=c_, linewidths=6, alpha=0.7)
plt.legend(results_6m.index)
plt.xlim(xmin=0, xmax=0.7)
plt.ylim(ymin=0, ymax=0.3)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Crash within 6 month - Recall vs Precision')
plt.grid()
plt.show()
```

```python
# Barplot to compare scores of models - 16 features
h6 = results_6m['score_t']
h3 = results_3m['score_t']
h1 = results_1m['score_t']
rcParams['figure.figsize'] = 11, 4.5
plt.bar(x=np.arange(len(h6))-0.2,width=0.2, height=h6, color='lightsalmon', label='6 months')
plt.bar(x=np.arange(len(h3)),width=0.2, height=h3, color='tomato', label='3 months')
plt.bar(x=np.arange(len(h1))+0.2, width=0.2, height=h1, color='brown', label='1 month')
plt.tight_layout()
plt.grid()
plt.legend()
plt.ylabel('F-beta score (beta=2)')
plt.xticks(np.arange(len(h6)), list(results_6m.index))
plt.title('F-beta scores for crash prediction in 6 / 3 / 1 months')
plt.show()
```
