## Predicting Crashes in Financial Markets -- Comparison of Results
- Linear Regression / Logistic Regression / SVM (linear) / SVM (nonlinear) / Decision Trees
- 16 features (for all models) vs 504 features (for regression models)
- Predictions for crash withinin 1 month / 3 months / 6 months

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
# 1 month prediction results:
d = {}
d['Linear Reg'] = [0.17, 0.3, 0.97, 0.26]
d['Logistic Reg'] = [0.1, 0.55, 0.91, 0.29]
d['SVC lin'] = [0.08, 0.50, 0.83, 0.25] #############
d['SVC RBF'] = [0.09, 0.51, 0.91, 0.26]
d['Decision Tree'] = [0.05, 0.1, 0.9, 0.1] #############
d['RNN LSTM'] = [0.08, 0.45, 0.90, 0.24] #############
d['random'] = [0.03, 0.1, 0.90, 0.05] #############
results_1m = pd.DataFrame.from_dict(d, orient='index')
results_1m.columns = ['precision_t', 'recall_t', 'accuracy_t', 'score_t']

# 3 months prediction results:
d = {}
d['Linear Reg'] = [0.15, 0.59, 0.84, 0.37]
d['Logistic Reg'] = [0.15, 0.56, 0.85, 0.36]
d['SVC lin'] = [0.16, 0.52, 0.87, 0.36]
d['SVC RBF'] = [0.16, 0.52, 0.87, 0.36]
d['Decision Tree'] = [0.1, 0.15, 0.95, 0.25] #############
d['RNN LSTM'] = [0.14, 0.55, 0.88, 0.35] #############
d['random'] = [0.03, 0.1, 0.90, 0.05] #############
results_3m = pd.DataFrame.from_dict(d, orient='index')
results_3m.columns = ['precision_t', 'recall_t', 'accuracy_t', 'score_t']

# 6 months prediction results:
d = {}
d['Linear Reg'] = [0.14, 0.68, 0.67, 0.38]
d['Logistic Reg'] = [0.16, 0.66, 0.72, 0.4]
d['SVC lin'] = [0.17, 0.6, 0.77, 0.4]
d['SVC RBF'] = [0.15, 0.71, 0.69, 0.41]
d['Decision Tree'] = [0.1, 0.15, 0.95, 0.25] #############
d['RNN LSTM'] = [0.14, 0.55, 0.88, 0.35] #############
d['random'] = [0.03, 0.1, 0.90, 0.05] #############
results_6m = pd.DataFrame.from_dict(d, orient='index')
results_6m.columns = ['precision_t', 'recall_t', 'accuracy_t', 'score_t']

# Plot results recall vs precision:
x = list(results_1m['recall_t'])
y = list(results_1m['precision_t'])
color = ['yellow', 'orange', 'blue', 'darkblue', 'green', 'red', 'lightgray']
marker=['o','o','o','o','o','o','o']
rcParams['figure.figsize'] = 8, 18
plt.subplot(3,1,1)
for x_, y_, c_, m_ in zip(x, y, color, marker):
    plt.scatter(x_, y_, marker=m_, c=c_, linewidths=6, alpha=0.7)
plt.legend(results_1m.index, loc='lower right')
plt.xlim(xmin=0, xmax=0.8)
plt.ylim(ymin=0, ymax=0.2)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Crash within 1 month - Recall vs Precision')
plt.grid()
plt.subplot(3,1,2)
x = list(results_3m['recall_t'])
y = list(results_3m['precision_t'])
for x_, y_, c_, m_ in zip(x, y, color, marker):
    plt.scatter(x_, y_, marker=m_, c=c_, linewidths=6, alpha=0.7)
plt.legend(results_3m.index, loc='lower right')
plt.xlim(xmin=0, xmax=0.8)
plt.ylim(ymin=0, ymax=0.2)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Crash within 3 month - Recall vs Precision')
plt.grid()
plt.subplot(3,1,3)
x = list(results_6m['recall_t'])
y = list(results_6m['precision_t'])
for x_, y_, c_, m_ in zip(x, y, color, marker):
    plt.scatter(x_, y_, marker=m_, c=c_, linewidths=6, alpha=0.7)
plt.legend(results_6m.index, loc='lower right')
plt.xlim(xmin=0, xmax=0.8)
plt.ylim(ymin=0, ymax=0.2)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Crash within 6 month - Recall vs Precision')
plt.grid()
plt.show()
```

```python
# 1 month prediction results:
d = {}
d['Linear Regression'] = [0.17, 0.3, 0.97, 0.26]
d['Logistic Regression'] = [0.1, 0.55, 0.91, 0.29]
d['SVC (linear)'] = [0.08, 0.50, 0.83, 0.25] #############
d['SVC (RBF)'] = [0.09, 0.51, 0.91, 0.26]
d['Decision Tree'] = [0.09, 0.08, 0.96, 0.12] #############
d['RNN LSTM'] = [0.14, 0.42, 0.93, 0.28] #############
d['random'] = [0.02, 0.04, 0.94, 0.033]
results_1m = pd.DataFrame.from_dict(d, orient='index')
results_1m.columns = ['precision_t', 'recall_t', 'accuracy_t', 'score_t']

# 3 months prediction results:
d = {}
d['Linear Regression'] = [0.15, 0.59, 0.84, 0.37]
d['Logistic Regression'] = [0.15, 0.56, 0.85, 0.36]
d['SVC (linear)'] = [0.16, 0.52, 0.87, 0.36]
d['SVC (RBF)'] = [0.16, 0.52, 0.87, 0.36]
d['Decision Tree'] = [0.09, 0.09, 0.91, 0.08] #############
d['RNN LSTM'] = [0.15, 0.5, 0.83, 0.32] #############
d['random'] = [0.04, 0.18, 0.79, 0.106]
results_3m = pd.DataFrame.from_dict(d, orient='index')
results_3m.columns = ['precision_t', 'recall_t', 'accuracy_t', 'score_t']

# 6 months prediction results:
d = {}
d['Linear Regression'] = [0.14, 0.68, 0.67, 0.38]
d['Logistic Regression'] = [0.16, 0.66, 0.72, 0.4]
d['SVC (linear)'] = [0.17, 0.6, 0.77, 0.4]
d['SVC (RBF)'] = [0.15, 0.71, 0.69, 0.41]
d['Decision Tree'] = [0.13, 0.15, 0.84, 0.14] #############
d['RNN LSTM'] = [0.15, 0.69, 0.63, 0.38] #############
d['random'] = [0.08, 0.35, 0.63, 0.209]
results_6m = pd.DataFrame.from_dict(d, orient='index')
results_6m.columns = ['precision_t', 'recall_t', 'accuracy_t', 'score_t']

colors = ['royalblue', 'blueviolet', 'gold', 'coral', 'green', 'red', 'gray']
markers = ['s', 'o', 'D']
f = lambda m,c: plt.plot([],[],marker=m, color=c, ls="none")[0]
handles = [f('s', colors[i]) for i in range(7)]
handles += [f(markers[i], 'k') for i in range(3)]
labels = list(results_6m.index) + ['1 month prediction', '3 month prediction', '6 month prediction']

# Plot results recall vs precision:
rcParams.update({'font.size': 17})
rcParams['figure.figsize'] = 11, 9
plt.style.use('seaborn-darkgrid')
x = list(results_1m['recall_t'])
y = list(results_1m['precision_t'])
for x_, y_, c_ in zip(x, y, colors):
    plt.scatter(x_, y_, marker='s', c=c_, linewidths=1, s=100, alpha=0.8)
x = list(results_3m['recall_t'])
y = list(results_3m['precision_t'])
for x_, y_, c_ in zip(x, y, colors):
    plt.scatter(x_, y_, marker='o', c=c_, linewidths=1, s=110, alpha=0.8)
x = list(results_6m['recall_t'])
y = list(results_6m['precision_t'])
for x_, y_, c_ in zip(x, y, colors):
    plt.scatter(x_, y_, marker='D', c=c_, linewidths=1, s=90, alpha=0.8)
plt.legend(handles, labels, loc=4, framealpha=1)
plt.xlim(xmin=0, xmax=1)
plt.ylim(ymin=0, ymax=0.2)
plt.xlabel('Recall')
plt.ylabel('Precision')
#plt.title('Recall vs Precision')
plt.show()
```

```python
rcParams['figure.figsize'] = 10, 10
rcParams.update({'font.size': 16})
df = pd.concat([results_6m['score_t'], results_3m['score_t'], results_1m['score_t']], axis=1)
df.columns = ['6 month prediciton', '3 month prediciton', '1 month prediciton']
df = df.iloc[::-1]
df.plot(kind='barh', stacked=False, width=0.75, color=['mediumaquamarine','mediumseagreen','green'])
plt.xlabel('F-beta score (beta = 2)')
plt.legend(loc='best', bbox_to_anchor=(1, 0.5))
plt.show()
```
